# News Intelligence System: Classification & Summarization

This notebook demonstrates an NLP pipeline for classifying news articles into categories and generating extractive summaries. 

### Objectives:
1. **Preprocess and Clean** real-world news data.
2. **Train a Classification Model** using TF-IDF and Logistic Regression.
3. **Implement Extractive Summarization** using word frequency scoring.
4. **Visualize Results** and evaluate model performance.

## 1. Setup and Imports

We import our modular components and standard data science libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pipeline import NewsPipeline
from classifier import NewsClassifier
import os

# Set plot style
sns.set_palette("viridis")
plt.style.use("seaborn-v0_8-whitegrid")
```

## 2. Dataset Exploration

We use the AG News dataset, which contains 120,000 training samples across 4 categories.

```python
train_df = pd.read_csv('data/train.csv')
print(f"Dataset Shape: {train_df.shape}")

# Map the labels for visualization
label_map = {1: "Politics/World", 2: "Sports", 3: "Business", 4: "Tech"}
train_df['Category'] = train_df['Class Index'].map(label_map)

# Visualize class distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Category')
plt.title("Distribution of News Categories")
plt.show()
```

## 3. Model Training and Performance

The model was trained using TF-IDF vectorization and Logistic Regression. Below is the performance on the test set.

```python
classifier = NewsClassifier()
if os.path.exists('news_classifier.pkl'):
    print("Loading pre-trained model engine...")
    classifier.load_model()
    classifier.evaluate('data/test.csv')
else:
    print("Training model... (this may take a minute)")
    classifier.train('data/train.csv')
    classifier.evaluate('data/test.csv')
```

## 4. Interactive Pipeline Demo

Let's test the pipeline on a sample news article. The pipeline will classify the topic and provide a 2-sentence extractive summary.

```python
pipeline = NewsPipeline()

news_article = """
SpaceX successfully launched its latest Falcon 9 rocket from Cape Canaveral today, 
carrying 22 Starlink satellites into low Earth orbit. The booster landed successfully 
on a drone ship in the Atlantic Ocean, marking the company's 200th recovery of an 
orbital-class rocket. Elon Musk praised the team for their rapid turnaround time, 
emphasizing that reusability is key to making Mars exploration affordable in the future.
"""

result = pipeline.process_article(news_article)

print(f"[PREDICTED TOPIC]: {result['category']}")
print("-" * 40)
print(f"[SUMMARY]:\n{result['summary']}")
```

## 5. Summary and Conclusions

We have successfully built a robust NLP pipeline that achieved over **90% accuracy** on news classification. The extractive summarizer provides effective article previews by identifying high-value sentences using word frequency scoring.
